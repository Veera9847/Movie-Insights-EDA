```python
import time
from selenium import webdriver
from bs4 import BeautifulSoup
import re

urls = ["https://www.imdb.com/list/ls562330283/", 
        "https://www.imdb.com/list/ls522454778/",
        "https://www.imdb.com/list/ls516348648/",
        "https://www.imdb.com/list/ls513706508/",
        "https://www.imdb.com/list/ls093492189/",
        "https://www.imdb.com/list/ls041278879/",
        "https://www.imdb.com/list/ls043581860/",
        "https://www.imdb.com/list/ls514198042/"]

driver = webdriver.Chrome()

all_movies = []
all_years = []
all_runtimes = []
all_ratings = []
all_directors = []
all_languages = []
all_budgets = []
all_first_day_collections = []
all_worldwide_collections = []
all_overseas_collections = []
all_india_gross_collections = []
all_verdicts = []

for url in urls:
    driver.get(url)
    
    time.sleep(3)  
    
    html = driver.page_source
    soup = BeautifulSoup(html, "html.parser")

    movies = []
    for i in soup.find_all("div", class_="ipc-title ipc-title--base ipc-title--title ipc-title-link-no-icon ipc-title--on-textPrimary sc-a69a4297-2 bqNXEn dli-title with-margin"):
        movies.append(i.text)

    for i in range(len(movies)):
        movies[i] = re.sub(r'^\d+\.\s*', '', movies[i])
    all_movies.extend(movies)

    l = []
    for i in soup.find_all("span", class_="sc-300a8231-7 eaXxft dli-title-metadata-item"):
        l.append(i.text)

    years = re.findall(r"\b(19\d{2}|20\d{2})\b", " ".join(l))  # Matches years like 1990, 2024, etc.
    all_years.extend(years)

    runtimes = re.findall(r"\d+h \d+m|\d+h", " ".join(l))
    all_runtimes.extend(runtimes)

    ratings = []
    for i in soup.find_all("span", class_="ipc-rating-star--rating"):
        ratings.append(i.text)
    all_ratings.extend(ratings)

    j=[]
    for i in soup.find_all("div", class_="sc-8c2fcf05-1 hOfyuG"):
        j.append(i.text)
    for k in j:
        director_match = re.search(r'Director(.*?)Stars', k)
        language_match = re.search(r'Language\s*:\s*(.*?)\s*Budget', k)
        budget_match = re.search(r'Budget\s*:\s*(.*?)\s*First Day Collection', k)
        first_day_collection_match = re.search(r'First Day Collection (?:India|Worldwide)\s*:\s*₹?([\d,]+(\.\d+)?cr)', k)
        worldwide_collection_match = re.search(r'Worldwide Collection\s*:\s*(.*?)\s*Overseas Collection', k)
        overseas_collection_match = re.search(r'Overseas Collection\s*:\s*(.*?)\s*India Gross Collection', k)
        india_gross_collection_match = re.search(r'India Gross Collection\s*:\s*(.*?)\s*Verdict', k)
        verdict_match = re.search(r"Verdict :\s*(.*?)(?:Note|$)", k)

        if director_match:
            all_directors.append(director_match.group(1))
        else:
            all_directors.append("Null")
        if language_match:
            all_languages.append(language_match.group(1))
        if budget_match:
            all_budgets.append(budget_match.group(1))
        if first_day_collection_match:
            all_first_day_collections.append(first_day_collection_match.group(1))
        else:
            all_first_day_collections.append("Null")
        if worldwide_collection_match:
            all_worldwide_collections.append(worldwide_collection_match.group(1))
        else:
            all_worldwide_collections.append("Null")
        if overseas_collection_match:
            all_overseas_collections.append(overseas_collection_match.group(1))
        else:
            all_overseas_collections.append("Null")
        if india_gross_collection_match:
            all_india_gross_collections.append(india_gross_collection_match.group(1))
        else:
            all_india_gross_collections.append("Null")
        if verdict_match:
            all_verdicts.append(verdict_match.group(1))
        else:
            all_verdicts.append("Null")

print("Total Movies:", len(all_movies))
print("Total Years Extracted:", len(all_years))
print("Total Run Times:", len(all_runtimes))
print("Total Ratings:", len(all_ratings))
print("Total Directors:", len(all_directors))
print("Total Languages:", len(all_languages))
print("Total Budgets:", len(all_budgets))
print("Total First Day Collections:", len(all_first_day_collections))
print("Total Worldwide Collections:", len(all_worldwide_collections))
print("Total Overseas Collections:", len(all_overseas_collections))
print("Total India Gross Collections:", len(all_india_gross_collections))
print("Total Verdicts:", len(all_verdicts))
```

```python
df =pd.DataFrame({'Movie_Name':all_movies,'year_of_release':all_years,'rating':all_ratings,'Run_time':all_runtimes,'Director':all_directors,'Language':all_languages,'Budget':all_budgets,'first_day_collection':all_first_day_collections,
              'Worldwide_collection':all_worldwide_collections,'Overseas_collection':all_overseas_collections,
             'India_gross_collection':all_india_gross_collections,'Verdict':all_verdicts})
df=df.sort_values(by='year_of_release',ascending=False)
df.to_csv('imdb.csv', index=False)
```

## Data Cleaning

```python
categories = ['All Time Blockbuster', 'Blockbuster', 'Super Hit', 'Hit', 'Above Average', 'Below Average','Average', 'Flop', 'Disaster', 'Below Average']

cleaned_verdicts=[]
for verdict in df['Verdict']:
    verdict_cleaned = "Average" 
    for category in categories:
        if re.search(rf"\b{category}\b", verdict, re.IGNORECASE):
            verdict_cleaned = category
            break
    cleaned_verdicts.append(verdict_cleaned)
print(cleaned_verdicts)
df['Verdict']=cleaned_verdicts
```
